```python
#importing CSV File
import pandas as pd
import numpy as np
results = pd.read_csv('MasterPlayerDataSet.csv')
```

```python
#Rank combine results to get a relative performance measure 

results['40 Rank'] = results['Forty'].rank(method='dense', ascending=True)
results['Vertical Rank'] = results['Vertical'].rank(method='dense', ascending=False)
results['BenchReps Rank'] = results['BenchReps'].rank(method='dense', ascending=False)
results['BroadJump Rank'] = results['BroadJump'].rank(method='dense', ascending=False)
results['Cone Rank'] = results['Cone'].rank(method='dense', ascending=True)
results['Shuttle Rank'] = results['Shuttle'].rank(method='dense', ascending=True)
results['average combine rank'] = results[['40 Rank', 'Shuttle Rank']].mean(axis=1)
results
```

```python
round = results.groupby(['Round']).mean()
round
```

```python
#Is combine performance a good indicator of Draft Selection?
%matplotlib inline 
import matplotlib.pyplot as plt
plt.scatter(round['Pick'], round['average combine rank'])
plt.show()
```

```python
results['Probowl#'] = results['Probowls']

ProbowlCount = results.groupby(['Probowl#']).mean()
ProbowlCount
```

```python
#Is combine performance a good indicator of ProBowl Selection?
%matplotlib inline 
import matplotlib.pyplot as plt
plt.scatter(ProbowlCount['Probowls'], ProbowlCount['average combine rank'])
plt.show()
```
